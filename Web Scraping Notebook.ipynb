### Web Scraping Steps:
- In this project, I have scraped the lastest data from: <br>
> - [CarGurus - A Canada's Automotive Marketplace](https://www.cargurus.ca)
- Repeatedly run the below code for several times to scrape the data using Selenium, and then saved the file and renamed it.
- After scraping enough data, all the files are merged into one dataset.


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time

# Create an empty DataFrame to store all vehical data
total_vehical_dataset = pd.DataFrame()

# Set up the web driver
driver = webdriver.Chrome()

try: 
    for run in range(1):

        vehical_data = []
     
        driver.get('https://www.cargurus.ca/Cars/l-Used-Tesla-m112#listing=395342775/NONE/DEFAULT')

        for page in range(700):
            data = {}
            try:
                title = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, 'h1[class="us1dS FmeUl _listingHeading_s8u01_6"]')))
                data["title"] = title.text
            except:
                    data["title"] = None
                    
            try:
                price = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, 'h2[class="us1dS CrxtQ"]')))
                data["price"] = price.text
            except:
                data["price"] = None
                
            try:
                location = WebDriverWait(driver, 5).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, 'p[class="us1dS iK3Zj"]')))
                data["location"] = location.text
            except:
                data["location"] = None
                
            try:
                spec_keys = WebDriverWait(driver, 10).until(
                    EC.presence_of_all_elements_located((By.CSS_SELECTOR,'h5[class="r-ugg _2C5uP"]')))
                spec_values = WebDriverWait(driver, 10).until(
                    EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'p[class="r-ugg"]')))
                
                print("Successfully located list 1")
                
                for key, value in zip(spec_keys, spec_values):
                    data[key.text.strip()] = value.text.strip()
            except:
                print("Error: Unable to locate list 1")
                
            try:
                overview_keys = WebDriverWait(driver, 10).until(
                    EC.presence_of_all_elements_located((By.CSS_SELECTOR,'span[class="us1dS iK3Zj _label_zbkq7_7"]')))
                overview_values = WebDriverWait(driver, 10).until(
                    EC.presence_of_all_elements_located((By.CSS_SELECTOR,'span[class="us1dS iK3Zj _value_zbkq7_14"]')))
                print("Successfully located list 2")
                
                for key, value in zip(overview_keys, overview_values):
                    data[key.text.strip()] = value.text.strip()
            except:
                print("Error: Unable to locate list 2")
        
            vehical_data.append(data)
    
            try: 
                time.sleep(1)
                next_button = driver.find_element(By.XPATH, '//*[@id="cargurus-listing-search"]/div/div[1]/div[2]/a[2]/span/span[2]')
                ActionChains(driver).move_to_element(next_button).click().perform()
                print("Successfully clicked 'Next' button.")
            except:
                print("No more pages or unable to locate 'Next' button.")
                break

        run_dataframe = pd.DataFrame(vehical_data)
        total_vehical_dataset = pd.concat([total_vehical_dataset, run_dataframe], ignore_index=True)

finally:
    driver.quit()

print(total_vehical_dataset.head())

total_vehical_dataset.to_csv('total_vehical_dataset.csv', index=False)
```
